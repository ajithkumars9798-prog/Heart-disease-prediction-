```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv("/content/heart.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Sex']=le.fit_transform(df['Sex'])
```

```python
df['ChestPainType']=le.fit_transform(df['ChestPainType'])
```

```python
df['RestingECG']=le.fit_transform(df['RestingECG'])
```

```python
df['ExerciseAngina']=le.fit_transform(df['ExerciseAngina'])
```

```python
df['ST_Slope']=le.fit_transform(df['ST_Slope'])
```

```python
df
```

```python
x=df.drop('HeartDisease',axis=1)
y=df['HeartDisease']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=42)
```

```python
lr=LogisticRegression(max_iter=1000)
lr.fit(x_train,y_train)
```

```python
y_prediction=lr.predict(x_train)
training_accuracy_score=accuracy_score(y_prediction,y_train)
print("training_accuracy_score",training_accuracy_score)
```

```python
y_test_prediction=lr.predict(x_test)
test_accuracy_score=accuracy_score(y_test_prediction,y_test)
print('test_accuracy_score',test_accuracy_score)
```
